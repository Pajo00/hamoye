```python
import pandas as pd

# Create a DataFrame from the given information
df = pd.DataFrame({
    "Year": [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    "Books Displayed": [100000, 120000, 140000, 160000, 180000, 200000, 220000, 240000, 260000, 280000, 300000],
    "Books Bought": [50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000, 140000, 150000],
    "Book Types": ["Fiction", "Non-Fiction", "Children's", "Science", "History", "Art", "Biography", "Travel", "Cookbooks", "Self-Help", "Romance"]
})

# Calculate the percentage of books bought for each type
df['Percentage Bought'] = df['Books Bought'] / df['Books Displayed'] * 100

# Sort the DataFrame by the percentage of books bought
df_sorted = df.sort_values(by='Percentage Bought', ascending=False)

# Get the top 3 frequently bought book types
top_3_types = df_sorted['Book Types'].head(3).tolist()

# Print the results
print("Top 3 frequently bought book types between 2009 and 2019:")
for type in top_3_types:
    print(f"- {type}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame from the given information
df = pd.DataFrame({
    "Year": [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    "Books Displayed": [100000, 120000, 140000, 160000, 180000, 200000, 220000, 240000, 260000, 280000, 300000],
    "Books Bought": [50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000, 140000, 150000],
    "Book Types": ["Fiction", "Non-Fiction", "Children's", "Science", "History", "Art", "Biography", "Travel", "Cookbooks", "Self-Help", "Romance"]
})

# Calculate the percentage of books bought for each type
df['Percentage Bought'] = df['Books Bought'] / df['Books Displayed'] * 100

# Sort the DataFrame by the percentage of books bought
df_sorted = df.sort_values(by='Percentage Bought', ascending=False)

# Get the top 3 frequently bought book types
top_3_types = df_sorted['Book Types'].head(3).tolist()

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_3_types, df_sorted['Percentage Bought'].head(3))
plt.xlabel("Book Type")
plt.ylabel("Percentage Bought (%)")
plt.title("Top 3 Most Frequently Bought Book Types (2009-2019)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
